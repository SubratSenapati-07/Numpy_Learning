### ðŸ”¹ ADDING AND REMOVING DATA :

```python
import numpy as np
```

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

# COMBINED :
combine = np.concatenate((arr1,arr2))
print("Combined array : ",combine)
```

#### Array compatibility :

```python
# Here ,we check shape of two or more array's is same or not.
a = np.array([4,5,6])
b = np.array([1,2,0])
c = np.array([4,1,2])
print("Compatibility shapes: ", a.shape == b.shape and b.shape == c.shape)
```

```python
org = np.array([[1,2],[2,3]])
new_row = np.array([[1,5]])

with_new_row = np.vstack((org,new_row)) # vstack se new row add hoti hai.
# print(org)
print("With new row: ",(with_new_row)) # merge in row.

new_col = np.array([[4,1],[2,8]])
print("With new column : ",np.hstack((org,new_col)) )
```

### DELETION :
