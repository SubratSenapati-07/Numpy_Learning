ARRAY'S PROPERTIES :

```python
import numpy as np
```

ðŸ”¹SYNTAX : array_name.properties   (shape,ndim,size,dtype etc)

```python
arr1 = np.array([[4,8,3],
                [7,0,4]])


## properties:

print("Shape: ",arr1.shape)
print("Dimension: ",arr1.ndim)
print("Size: ",arr1.size)
print("Data Type: ",arr1.dtype)
```

ðŸ”¹Array Reshaping 

```python
# NORMAL ARRAY : 
arr2 = np.arange(12)   # arange(parameter) # arange(start,end)
print("Original array: ",arr2)

# RESHAPING :
reshaped = arr2.reshape((3,4)) # reshape((row,column))
print("Reshaped array : \n",reshaped)

# Flatting i.e. In a Single Row :
flat = reshaped.flatten()
print("\n Flattened array: ",flat)

# Ravele : (returns view , instead of copy)
ravele = flat.ravel()
print("\n reveled array:",ravele)

# Transpose :
transpose = reshaped.T
print("Transpose: \n",transpose)
```
