```python
import numpy as np

arr1 = np.array([1,5,50])
arr2 = np.array([2,7,0])
sum = arr1 + arr2
print("Sum of two arrays: ",sum)
```

#### 2D Arrays :

```python
arr01 = np.array([[1,10],
                  [2,11]])
arr02 = np.array([[2,0],
                  [9,1]])
sumof = arr01 + arr02
print("Sum of 2D arrays:\n",sumof)
```
