CREATING ARRAY FROM SCRATCH / METHODS

```python
import numpy as np
```

ðŸ”¹zeros gives null matrix, Size is
 terminated 
   within tuple (row,column).
   e.g - (3,4)

```python
zeros = np.zeros((3,4))
print("Zeros array: \n",zeros)
```

ðŸ”¹ones gives unit matrix, Size is
 terminated 
   within tuple (row,column).
   e.g - (2,3)

```python
one = np.ones((2,3))
print("unit matrix:\n",one)
```

ðŸ”¹full gives full same element's matrix, Size is
 terminated 
   within tuple ((row,column),element ).
   e.g - ((3,3),5)

```python
full = np.full((3,3), 5) 
print("full array:\n",full)
```

ðŸ”¹random : for random values matrix. (row,colum)

ðŸ”¹Values are fluetuate within 0 - 1

```python
random = np.random.random((2,2))
print("Random matrix:\n",random)
```

ðŸ”¹sequence : arange(start,end,stepSize) AND it gives a sequence array.

```python
seq = np.arange(0,11)
print("Sequence array:",seq)

# OR
seq1 = np.arange(0,11,2)
print("Seq1 array:",seq1)
```
