Vector,Matrix And Tensor

```python
import numpy as np
```

```python
vector = np.array([5,8,6])
print("Vector: ",vector)

matrix =np.array([[1,2,0],
                  [7,0,1]])
print("Matrix:\n",matrix)

tensor = np.array([[[1,2],
                    [4,5]],
                   [[1,8],
                    [7,6]]])
print("Tensor:\n",tensor)
```
