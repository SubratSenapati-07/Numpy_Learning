### ðŸ”¹ VECTOR OPERATION 

```python
import numpy as np
```

```python
vector1 = np.array([2,5,6,7,8])
vector2 = np.array([7,9,3,1,7])

print("Vector addition : ",vector1 + vector2)
print("Vector multiplication : ",vector1 * vector2)
print("Dot product : ", np.dot(vector1 , vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))

print("Angle (radians):", angle)
print(f"Angle (degrees): {round(np.degrees(angle), 3)}Â°")
```

```python
#VECTORIZING: 

res_typ = np.array(["Pasta Wine","Libya","EX Ero","Biriyani","Royal Mint"])
vectorized = np.vectorize(str.upper)
print("Vectorized value: ",vectorized(res_typ))
```
