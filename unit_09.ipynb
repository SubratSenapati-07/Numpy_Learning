### ðŸ”¹ADVANCE OPERATION WITH BUSINESS EXAMPLE :

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure : [ restaurant_id, 2022 , 2023, 2024, 2025]

sales_data = np.array([
    [1, 150000 , 180000, 220000, 250000 ], # Desi Biriyani
    [2, 120000, 140000, 160000, 190000],     # Foreign res
    [3, 200000, 230000, 260000, 300000],     # Pizza Store
    [4, 180000, 210000, 240000, 270000],     # Burger point
    [5, 160000,185000 ,205000, 230000]     # Dolly Chai
])

print("==== Zomato Sales analysis ====")
print("\n Sales data shape :",sales_data.shape)
print("\n Sample data for 1st 3 restaurant:\n",sales_data[0:3, 1:4])  # [Column slicing , Row slicing]
```

```python
# Total Sale per year: 
print(np.sum(sales_data,axis=0),"\n")

# We erase first column :
print("Total sales per year:",np.sum(sales_data[0:5,1:5],axis=0))

# axis = 0 : add individual row. 
# axis = 1 : add individual data from column.


yearly = np.sum(sales_data[0:5,1:5],axis=1)
print("All Restaurant total yearly data  :",np.sum(yearly))

# Minimum sales per restaurant :
min_sale = np.min(sales_data[:,1:],axis = 1)
print("Minimum sales per restaurant:",min_sale)

# Maximum sales per year :
max_sale = np.max(sales_data[0:5,1:5],axis=0)
print(max_sale)


# Average sale per restaurant :
avg_sale = np.mean(sales_data[:,1:],axis=1)

# Cumulative Sum :
cusum = np.cumsum(sales_data[:,1:],axis=1)
print(cusum)
```

### ðŸ”¹ Graph Plote :

```python
plt.figure(figsize = (5, 4))
plt.plot(np.mean(cusum, axis = 0))
plt.title("Average Cumalative sales across all restaurant: ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
